```python

from allennlp.predictors.predictor import Predictor
from allennlp_models.pretrained import get_pretrained_models
from allennlp_models.pretrained import load_predictor
import logging

import nltk.data

logging.disable(logging.CRITICAL)

with open("resources/texts/Germany.txt", 'r') as file:
        loaded = file.read()

tokenizer = nltk.data.load('nltk:tokenizers/punkt/english.pickle')
tokenized = tokenizer.tokenize(loaded)


predictor = load_predictor("structured-prediction-srl-bert")

global predictions
predictions = []

for token in tokenized:
    predictions.append(predictor.predict(token))

predictions[0]
```
